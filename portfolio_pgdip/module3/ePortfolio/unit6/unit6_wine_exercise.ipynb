```python
# initial setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

import sklearn
import random 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler

```

```python
# import datasets

data=pd.read_csv("Datasets\\Unit06 wine.csv")
```

```python
# general exploration

data.info()

data.head()
```

```python
data.Wine.value_counts()
```

3 different types included. We'll remove this label before clustering. All other features are numeric and can therefore be included in principle

```python
# standardisation

## transform dataframe into array
array_clustering = data.drop('Wine', axis=1).values
array_clustering = np.nan_to_num(array_clustering)
array_clustering

## apply standardisation
array_clustering_standard = StandardScaler().fit_transform(array_clustering)
array_clustering_standard
```

```python
# apply clustering
k = 3
k_means = KMeans(init = "k-means++", n_clusters = k, n_init = 12)
k_means.fit(array_clustering_standard)
labels = k_means.labels_
print(labels)
```

```python
# bring cluster labels to dataset

data["cluster"] = labels
data.head(5)
```

```python
# assign cluster as factor

data.cluster=data['cluster'].astype(object)


data.info()
```

```python
# cross check clusters with species


pd.crosstab(data.Wine, data.cluster, margins=True)
```

Reasonable separation achieved into the different wines. Some overlap between 1 and 2.

```python
# Checking centroids

data.drop('Wine', axis=1).groupby('cluster').mean()
```
