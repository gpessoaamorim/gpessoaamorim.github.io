```python
# initial setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

import sklearn
import random 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler

```

```python
# import datasets

data=pd.read_csv("Datasets\\Unit06 iris.csv")
```

```python
# general exploration

data.info()

data.head()
```

Data includes a categorical variable (species), which needs to be removed before we run clustering.

```python
# data preparation

data_clustering = data.drop('species', axis=1)
```

```python
# standardisation

## transform dataframe into array
array_clustering = data_clustering.values
array_clustering = np.nan_to_num(array_clustering)
array_clustering

## apply standardisation
array_clustering_standard = StandardScaler().fit_transform(array_clustering)
array_clustering_standard
```

```python
# standardisation

## transform dataframe into array
array_clustering = data_clustering.values
array_clustering = np.nan_to_num(array_clustering)
array_clustering

## apply standardisation
array_clustering_standard = StandardScaler().fit_transform(array_clustering)
array_clustering_standard
```

```python
from sklearn.cluster import KMeans
k = 3
k_means = KMeans(init = "k-means++", n_clusters = k, n_init = 12)
k_means.fit(array_clustering_standard)
labels = k_means.labels_
print(labels)
```

```python
# bring cluster labels to dataset

data["cluster"] = labels
data.head(5)
```

```python
# assign cluster as factor

data.cluster=data['cluster'].astype(object)


data.info()
```

```python
# cross check clusters with species


pd.crosstab(data.species, data.cluster, margins=True)
```

Reasonable separation achieved into the different species. Some overlap between versicolor and virginica.

```python
# visualisation

# assign cluster labels
conditions = [
    (data['species'] == 'setosa'),
    (data['species'] == 'versicolor') & (data['cluster'] == 0),  # Use parentheses for each condition
    (data['species'] == 'versicolor') & (data['cluster'] == 2),
    (data['species'] == 'virginica') & (data['cluster'] == 0),
    (data['species'] == 'virginica') & (data['cluster'] == 2)
]

choices = ['setosa', 'versicolor0', 'versicolor2', 'virginica0', 'virginica2']

# Apply np.select
data['cluster_label'] = np.select(conditions, choices)  # Create a new column

# Display the updated DataFrame
data.head(100)
```

```python
# visualisation


ax1=sns.scatterplot(data, x='sepal_length', y='sepal_width', hue='cluster_label')
ax1.set_title("Sepal features by species")
ax1

```
